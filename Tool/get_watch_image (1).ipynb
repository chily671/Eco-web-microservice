```python
from icrawler.builtin import GoogleImageCrawler


folder_path = "D:\\Workspace\\Eco-web-microservice\\Tool\\watch_images"

google_Crawler = GoogleImageCrawler(
    storage={
        "root_dir": "D:\\Workspace\\Eco-web-microservice\\Tool\\watch_images"
    }
)
```

```python
import pandas as pd

watch_data = pd.read_csv("Watches.csv", index_col=None)
```

```python
# Thc hin ly hnh nh ng h v t tn file theo tn ca ng h
import time
import os

for i in range(len(watch_data)):
    google_Crawler.crawl(keyword=watch_data["name"][i], max_num=1)
    # i tn file 0001.png thnh INDEX ca ng h

    # Th mc cha cc file

    # Ly danh sch cc file trong th mc
    files = os.listdir(folder_path)

    # i tn cc file
    for file in files:
        if file.startswith("000001."):
            new_name = file.replace("000001.", f"{i}.")
            os.rename(
                os.path.join(folder_path, file), os.path.join(folder_path, new_name)
            )
```
